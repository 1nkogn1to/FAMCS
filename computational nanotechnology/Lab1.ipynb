Задание 1: На отрезке [2; 3] найти минимальное значение функции $$F(\alpha) = \int_{1}^{2}\Gamma (\alpha x) exp(- \alpha x) dx$$


Задание 2: Решить на отрезке [0; 2] дифференциальное уравнение третьего порядка,
приведя его к системе уравнений первого порядка. Вывести график решения
$$y^{(3)} + y^{(2)} - 10y^{(1)} + 9y = 0$$

```python
from scipy.special import gamma
from scipy.integrate import quad, solve_ivp
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt
import numpy as np


#Задание 1
def F(a):
    return quad(lambda x: gamma(a * x) * np.exp(-a * x), 1, 2)[0]

def task1():
    result = minimize_scalar(F, bounds=(2, 3))
    print(f"Минимальное значение функции на отрезке [2, 3] в точке {result.x}")

#Задание 2
def system(t, y):
    y1, y2, y3 = y
    return [y2, y3, -y3 + 10 * y2 - 9 * y1]

def task2():
    interval = [0, 2]
    start_conditions = [0, 0, 1]
    solution = solve_ivp(system, interval, start_conditions, t_eval=np.linspace(0, 2, 100))
    t = solution.t
    y1, y2, y3 = solution.y

    plt.plot(t, y1, label="y(t)")
    plt.plot(t, y2, label="y'(t)")
    plt.plot(t, y3, label="y''(t)")
    plt.legend()
    plt.show()

def main():
    task1()
    task2()

if __name__ == "__main__":
    main()
```
