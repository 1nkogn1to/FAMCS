Задание:
Решить задачу для стационарного уравнения Шредингера с помощью
вариационного метода, используя указанные в варианте пробные функции.
Сравнить энергии основного состояния, полученные с помощью разных пробных
функций. Указать, какие функции дают более точный результат. Для обеих функций
построить графики волновых функций основного состояния.
Для второй пробной функции на графике потенциала построить полученные уровни
энергии, а также волновые функции основного, первого и второго возбужденных
состояний.
$$-\psi''+V(x)\psi=E\psi$$
$$V(x)=-\frac{1}{\sqrt{x^2+1}}$$
$$\psi \rightarrow 0, при |x| \rightarrow \infty$$
$$1)\phi(x)=e^{-ax^2}$$
$$2)\phi_k(x)=x^ke^{-|x|}, k=0,1,...,9$$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as s
from scipy.optimize import minimize

x = s.symbols('x')
a = s.symbols('a')
phi = s.exp(-a * x**2)

diff = -s.diff(phi, x, 2) + 1 / s.sqrt(x**2 + 1) * phi

integral = s.integrate(diff, (x, -np.inf, np.inf), manual=True)
f = s.lambdify(a, integral, 'scipy')

fVal = minimize(f, 1.0, bounds=[(-np.inf, np.inf)]).fun
aVal = minimize(f, 1.0, bounds=[(-np.inf, np.inf)]).x[0]

print(f"{fVal} - энергия основного состояния")

phi = s.lambdify(x, phi.subs(a, aVal), 'numpy')

t = np.linspace(-0.1, 0.1, 1000)
phiVals = phi(t)

plt.plot(t, phiVals)
plt.xlabel('x')
plt.show()
```
